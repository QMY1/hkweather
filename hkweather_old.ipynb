```python
precipitation['month'] = pd.to_datetime(precipitation.index).month
precipitation['year'] = pd.to_datetime(precipitation.index).year
sns.scatterplot(x='month',y='Astore',hue='year',data=precipitation)
```

```python
sns.pairplot(data=precipitation,
             vars = ["Astore", "Bunji", "Chilas", "Dainyour", "Gilgit", "Hunza", "Khunjrab", "Ramma", "Shigar", "Skardu", "Ziarat"],  # names of columns used for the plot
             kind="scatter",                        # kind of plots for each pair of different columns
             diag_kind="kde",                       # kind of plots on the diagonal
             #hue="COLUMN",
             height=2,
             aspect=1.1,
             palette="muted"
             )

plt.suptitle("Pair plot", y=1.02)   # adds title to the plot
plt.show()
```

```python
#define grid
g = sns.FacetGrid(data=precipitation, col='Astore', col_wrap=3)

#add plots to grid
g.map(sns.scatterplot, "Bunji", "Chilas", "Dainyour", "Gilgit", "Hunza", "Khunjrab", "Ramma", "Shigar", "Skardu", "Ziarat")
```
